```r
for (i in 1:length(colnames(df))) {
    feature = colnames(df)[i]
    
    var.name = paste("exp", feature, sep="_")
    df[,var.name] <- exp(df[,feature])
    
    var.name = paste("log", feature, sep="_")
    df[,var.name] <- log(df[,feature])
    # df[!is.finite ( df[,var.name] ),] <- 0
    
    var.name = paste("x2", feature, sep="_")
    df[,var.name] <- (df[,feature])^2
    
    var.name = paste("tanh", feature, sep="_")
    df[,var.name] <- tanh(df[,feature])
}
```
