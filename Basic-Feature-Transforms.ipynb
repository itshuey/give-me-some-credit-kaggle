## Basic Bivariate Feature Transforms

```r
df <- read.csv("cs-training.csv")
df <- na.omit(df)
head(df)
```

```r
#df$MonthlyIncome <- df$MonthlyIncome / 1000
```

```r
head(df[7:length(df)])
```

```r
colnames(df)
```

```r
features <- colnames(df)
for (i in 4:length(features)) {
    for (j in 4:length(features)) {
        feature.one = colnames(df)[i]
        feature.two = colnames(df)[j]
        
        if ( !( is.na ( max(df[,feature.one]) * max(df[,feature.two]) ) ) ) {
            var.name = paste(feature.one, feature.two, sep="*")
            df[,var.name] <- df[,feature.one] * df[,feature.two]
        }

        var.name = paste(feature.one, feature.two, sep="+")
        df[,var.name] <- df[,feature.one] + df[,feature.two]
    }
}
```

```r
colnames ( df )
```

```r
corr <- na.omit ( as.data.frame(cor(na.omit(df)))[,2, drop=FALSE] )#[order(df$SeriousDlqin2yrs),]
```
